```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(a):
    exp = np.exp(a)
    return exp/sum(exp)
```

```python
def plot(x):
    for series in x:
        plt.plot(series)
        
    plt.show()
```

```python
a = [1, 2, 3, 4, 5]
s = softmax(a)

plot([a, s])
```

```python
a = np.random.randint(5, size=(5))
s = softmax(a)

plot([a, s])
```

```python
a = np.random.rand(5)
s = softmax(a)

plot([a, s])
```
