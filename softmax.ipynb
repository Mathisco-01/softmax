```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(a):
    exp = np.exp(a)
    return exp/sum(exp)
```

```python
def plot(x):
    count = 1
    color = {1: "red", 2: "blue", 3:"green"}
    for series in x:
        plt.plot(series, color=color[count])
        count += 1
        
    plt.show()
```

```python
a = [1, 2, 3, 4, 5]
s = softmax(a)
one = [sum(s)]*len(s) # sum of softmax probabilities

plot([a, s, one])
```

```python
a = np.random.randint(5, size=(5))
s = softmax(a)
one = [sum(s)]*len(s)

plot([a, s, one])
```
